```python
import pandas as pd

df = pd.read_csv('AmesHousing.csv')

df
```

```python
df.describe()
```

```python
df.info()
```

```python
#eliminar columnas con mas del 50% de valores nulos

porcentaje = 0.5
porcentaje_de_nulos = df.isnull().mean()

borrar = porcentaje_de_nulos[ porcentaje_de_nulos > porcentaje ].index.tolist()

df = df.drop(columns=borrar)

df.info()
```

```python
#eliminar columnas donde mas del 95% de las variables sean iguales
porcentaje = 0.95
baja_varianza = []

for columna in df.select_dtypes(include=['object']).columns:
  x = df[columna].value_counts(normalize=True).iloc[0]
  if x > porcentaje:
    baja_varianza.append(columna)

df = df.drop(columns=baja_varianza)
df.info()
```

```python
#selecciono las 15 variables que mas relacion pueden tener con la varianza del precio
variables_finales = ['Overall Qual','Gr Liv Area','Neighborhood','Total Bsmt SF','Garage Cars','1st Flr SF','Full Bath','Lot Area','Year Built','Year Remod/Add','Kitchen Qual','Foundation','Fireplaces','Mas Vnr Area','Exter Qual','SalePrice']

df = df[variables_finales]
df
```

```python
df.info()
```

Neighborhood --> dummies
Kitchen Qual --> one-hot
Foundation --> one-hot
Exter Qual --> one-hot

```python
# The 'Foundation' column will be one-hot encoded in a later step along with other categorical features.
```

```python
df = pd.get_dummies(df, columns=['Kitchen Qual', 'Exter Qual', 'Foundation', 'Neighborhood'], drop_first=True)
df.info()
```

```python
# Lista de todas las columnas booleanas para convertir a INT
columnas_bool = df.select_dtypes(include='bool').columns
df[columnas_bool] = df[columnas_bool].astype(int)

df.info()
df = df.fillna(0)
```

## Entrenamiento de Modelos

#### Regresion Lineal

```python
import numpy as np

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score


#Definicion de variables
X = df.drop('SalePrice', axis=1)
y = df['SalePrice']


#Conjuntos de entrenamiento
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

print(f"Número de características (k): {X_train.shape[1]}")
print(f"Número de observaciones de prueba (n): {X_test.shape[0]}")
```

```python
#defino y entreno el modelo
model = LinearRegression()
model.fit(X_train, y_train)

#predicciones
y_pred = model.predict(X_test)
```

```python
# 3. Cálculo de Métricas de Error
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)

print("--- Métricas de Error (Unidad: Cientos de Miles de Dólares) ---")
print(f"MAE (Error Absoluto Medio): {mae:.4f}")
print(f"MSE (Error Cuadrático Medio): {mse:.4f}")
print(f"RMSE (Raíz del Error Cuadrático Medio): {rmse:.4f}")
```

MAE ($20,181): De media, tu modelo se equivoca por unos 20 mil dólares. Es una cifra decente para un primer modelo.

RMSE ($32,063):Fijate que el RMSE es mucho mayor que el MAE. Esto te esta gritando que tienes outliers (valores atípicos).
El RMSE penaliza los errores grandes de forma cuadrática; si tienes un par de casas donde el modelo falló por 100.000, el RMSE subirá muchísimo aunque el MAE sea bajo.

```python
#r2score
rscore = r2_score(y_test,y_pred)
print(f"R2 Score del Modelo Base: {rscore:.4f}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,6))

sns.scatterplot(x = y_test, y=y_pred, alpha=0.5)

linea_prediccion_per = [y_test.min(), y_test.max()]
plt.plot(linea_prediccion_per, linea_prediccion_per, color='red', linestyle='--')

plt.title('Valores Reales vs. Predicciones (Ames Housing)')
plt.xlabel('Precio Real')
plt.ylabel('Precio Prediccion')
plt.show()
```

El RMSE es mucho más grande que mi MAE, lo que indica que el modelo comete errores muy grandes para ciertas casas. Es decir, para las casas con un precio que este dentro de la media, el error de prediccion es bajo, pero cuando se trata de casas muy baratas o muy caras, el modelo falla considerablemente.


Vamos a aplicar una escala logaritmica al precio, y entrenar el modelo con este cambio.

```python
#Definicion de variables
X = df.drop('SalePrice', axis=1)
y = np.log1p(df['SalePrice'])


#Conjuntos de entrenamiento
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

print(f"Número de características (k): {X_train.shape[1]}")
print(f"Número de observaciones de prueba (n): {X_test.shape[0]}")
```

```python
#defino y entreno el modelo
model = LinearRegression()
model.fit(X_train, y_train)

#predicciones
y_pred = model.predict(X_test)
```

```python
# 3. Cálculo de Métricas de Error
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)

# 4. Transformación INVERSA para volver a euros
pred_final = np.expm1(y_pred)
y_test_real = np.expm1(y_test)

# 5. Cálculo de métricas finales en EUROS
mae_final = mean_absolute_error(y_test_real, pred_final)
rmse_final = np.sqrt(mean_squared_error(y_test_real, pred_final))

print("--- Métricas de Error (Unidad: Cientos de Miles de Dólares) ---")
print(f"MAE (Error Absoluto Medio): {mae_final:.4f}")
print(f"MSE (Error Cuadrático Medio): {mse:.4f}")
print(f"RMSE (Raíz del Error Cuadrático Medio): {rmse_final:.4f}")
```

### Modelo con LASSO Regresion

```python
from sklearn.linear_model import LassoCV
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline

lasso_model = make_pipeline(StandardScaler(), LassoCV(cv=5, random_state=42))

# 2. Entrenamos (recuerda que y_train_log ya tiene aplicado el np.log1p)
lasso_model.fit(X_train, y_train)

# 3. Predicciones en escala logarítmica
pred_log = lasso_model.predict(X_test)

# 4. Transformación INVERSA para volver a euros
pred_final = np.expm1(pred_log)
y_test_real = np.expm1(y_test)

# 5. Cálculo de métricas finales en EUROS
mae_final = mean_absolute_error(y_test_real, pred_final)
rmse_final = np.sqrt(mean_squared_error(y_test_real, pred_final))

print(f"Mejor Alpha encontrado: {lasso_model.named_steps['lassocv'].alpha_}")
print(f"MAE final en euros: {mae_final:.2f}")
print(f"RMSE final en euros: {rmse_final:.2f}")
```

1. ¿Ha mejorado el modelo?
MAE anterior: 19.274,84 €

MAE actual (Lasso): 18.205,83 €

Conclusión: ¡Sí! Has reducido el error medio en unos 1.069 €. Aunque parezca poco, en modelos de regresión ganar un 5.5% de precisión solo cambiando el algoritmo es un éxito rotundo.

2. El misterio del RMSE
RMSE anterior: 30.712,12 €

RMSE actual (Lasso): 31.136,11 €

Análisis: Curiosamente, tu RMSE ha subido ligeramente. Esto sucede porque Lasso prefiere ser más preciso con la "gran mayoría" de las casas (bajando el MAE) a costa de no esforzarse tanto en los valores ultra-extremos. Lasso ha decidido que es mejor acertar bien 99 casas normales que volverse loco intentando predecir una mansión única.

3. Interpretación del Alpha
Alpha: 0.000328

Qué significa: Al ser un valor muy cercano a cero, Lasso te está diciendo que la mayoría de tus variables son útiles. No ha necesitado "borrar" demasiadas columnas para que el modelo funcione. Es un alpha "quirúrgico": solo elimina el ruido muy sutil.

Diagnóstico Final del Modelo
Tienes lo que en ciencia de datos llamamos un problema de Outliers persistentes.

Tu modelo es muy robusto para casas de precio estándar (MAE bajo).

Sin embargo, el RMSE alto indica que todavía hay algunas casas en tu set de test cuyos precios son impredecibles con las variables actuales (quizás casas con características de lujo muy específicas que no están en el dataset).
