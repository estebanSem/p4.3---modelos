```python
import pandas as pd

df = pd.read_csv('./data/AmesHousing.csv')

df
```

# Preprocesamiento y Limpieza de Datos

```python
df.describe()
```

```python
df.info()
```

```python
#eliminar columnas con mas del 50% de valores nulos

porcentaje = 0.5
porcentaje_de_nulos = df.isnull().mean()

borrar = porcentaje_de_nulos[ porcentaje_de_nulos > porcentaje ].index.tolist()

df = df.drop(columns=borrar)

df.info()
```

```python
#eliminar columnas donde mas del 95% de las variables sean iguales
porcentaje = 0.95
baja_varianza = []

for columna in df.select_dtypes(include=['object']).columns:
  x = df[columna].value_counts(normalize=True).iloc[0]
  if x > porcentaje:
    baja_varianza.append(columna)

df = df.drop(columns=baja_varianza)
df.info()
```

### Seleccion de 15 variables

```python
#selecciono las 15 variables que mas relacion pueden tener con la varianza del precio
variables_finales = ['Overall Qual','Gr Liv Area','Neighborhood','Total Bsmt SF','Garage Cars','1st Flr SF','Full Bath','Lot Area','Year Built','Year Remod/Add','Kitchen Qual','Foundation','Fireplaces','Mas Vnr Area','Exter Qual','SalePrice']

df_train = df[variables_finales]
df_train
```

```python
df_train.info()
```

```python
df_train = pd.get_dummies(df_train, columns=['Kitchen Qual', 'Exter Qual', 'Foundation', 'Neighborhood'], drop_first=True)
df_train.info()
```

```python
# Lista de todas las columnas booleanas para convertir a INT
columnas_bool = df_train.select_dtypes(include='bool').columns
df_train[columnas_bool] = df_train[columnas_bool].astype(int)

df_train.info()
df_train = df_train.fillna(0)
```

```python
df_train
```

# Entrenamiento de Modelos
## 15 Variables Clave

#### REGRESIN LINEAL

```python
import numpy as np

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score


#Definicion de variables
X = df_train.drop('SalePrice', axis=1)
y = df_train['SalePrice']


#Conjuntos de entrenamiento
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

print(f"Nmero de caractersticas (k): {X_train.shape[1]}")
print(f"Nmero de observaciones de prueba (n): {X_test.shape[0]}")
```

```python
#defino y entreno el modelo
model = LinearRegression()
model.fit(X_train, y_train)

#predicciones
y_pred = model.predict(X_test)
```

```python
# 3. Clculo de Mtricas de Error
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)

print("--- Mtricas de Error (Unidad: Cientos de Miles de Dlares) ---")
print(f"MAE (Error Absoluto Medio): {mae:.4f}")
print(f"MSE (Error Cuadrtico Medio): {mse:.4f}")
print(f"RMSE (Raz del Error Cuadrtico Medio): {rmse:.4f}")

#r2score
rscore = r2_score(y_test,y_pred)

n = X_test.shape[0]
k = X_test.shape[1]

#formula r2 ajustado
r2_ajustado = 1 - ((1 - rscore) * (n - 1) / (n - k - 1))

print(f"R2 Score : {rscore:.4f}")
print(f"R2 Score Ajustado: {r2_ajustado:.4f}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,6))

sns.scatterplot(x = y_test, y=y_pred, alpha=0.5)

linea_prediccion_per = [y_test.min(), y_test.max()]
plt.plot(linea_prediccion_per, linea_prediccion_per, color='red', linestyle='--')

plt.title('Valores Reales vs. Predicciones (Ames Housing)')
plt.xlabel('Precio Real')
plt.ylabel('Precio Prediccion')
plt.show()
```

#### REGRESION LINEAL CON ESCALA LOGARITMICA

### ELIMINACION DE OUTLIERS CON ISOLATION FOREST

```python
from sklearn.ensemble import IsolationForest

#aunque todas las columnas son numericas, es buena practica asegurarnos que no se cuela nada 
df_num = df_train.select_dtypes(include=[np.number])

# contamination es el % de datos que creo que son outliers
iso = IsolationForest(contamination=0.02, random_state=42)
preds = iso.fit_predict(df_num)

df_filtrado_iso = df_train[preds == 1]

print(f"Filas eliminadas con Isolation Forest: {len(df_train) - len(df_filtrado_iso)}")
```

### REGRESION LASSO

```python
from sklearn.linear_model import LassoCV
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline

#Definicion de variables
X = df_filtrado_iso.drop('SalePrice', axis=1)
y = np.log1p(df_filtrado_iso['SalePrice'])


#Conjuntos de entrenamiento
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

lasso_model = make_pipeline(StandardScaler(), LassoCV(cv=5, random_state=42))

# 2. Entrenamos (recuerda que y_train_log ya tiene aplicado el np.log1p)
lasso_model.fit(X_train, y_train)

# 3. Predicciones en escala logartmica
pred_log = lasso_model.predict(X_test)

# 4. Transformacin INVERSA para volver a euros
pred_final = np.expm1(pred_log)
y_test_real = np.expm1(y_test)

# 5. Clculo de mtricas finales en EUROS
mae_final = mean_absolute_error(y_test_real, pred_final)
rmse_final = np.sqrt(mean_squared_error(y_test_real, pred_final))

print(f"Mejor Alpha encontrado: {lasso_model.named_steps['lassocv'].alpha_}")
print(f"MAE final en euros: {mae_final:.2f}")
print(f"RMSE final en euros: {rmse_final:.2f}")

#r2score
rscore = r2_score(y_test_real,pred_final)

n = X_test.shape[0]
k = X_test.shape[1]

#formula r2 ajustado
r2_ajustado = 1 - ((1 - rscore) * (n - 1) / (n - k - 1))

print(f"R2 Score : {rscore:.4f}")
print(f"R2 Score Ajustado: {r2_ajustado:.4f}")
```

```python
plt.figure(figsize=(10,6))

sns.scatterplot(x = y_test_real, y=pred_final, alpha=0.5)

linea_prediccion_per = [y_test_real.min(), y_test_real.max()]
plt.plot(linea_prediccion_per, linea_prediccion_per, color='red', linestyle='--')

plt.title('Modelo LASSO')
plt.xlabel('Precio Real')
plt.ylabel('Precio Prediccion')
plt.show()
```
