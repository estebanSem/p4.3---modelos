```python
import pandas as pd

df = pd.read_csv('./data/AmesHousing.csv')

df
```

```python
df.describe()
```

```python
df.info()
```

```python
#eliminar columnas con mas del 50% de valores nulos

porcentaje = 0.5
porcentaje_de_nulos = df.isnull().mean()

borrar = porcentaje_de_nulos[ porcentaje_de_nulos > porcentaje ].index.tolist()

df = df.drop(columns=borrar)

df.info()
```

```python
#eliminar columnas donde mas del 95% de las variables sean iguales
porcentaje = 0.95
baja_varianza = []

for columna in df.select_dtypes(include=['object']).columns:
  x = df[columna].value_counts(normalize=True).iloc[0]
  if x > porcentaje:
    baja_varianza.append(columna)
    
df = df.drop(columns=baja_varianza)
df.info()
```

```python
#selecciono las 15 variables que mas relacion pueden tener con la varianza del precio
variables_finales = ['Overall Qual','Gr Liv Area','Neighborhood','Total Bsmt SF','Garage Cars','1st Flr SF','Full Bath','Lot Area','Year Built','Year Remod/Add','Kitchen Qual','Foundation','Fireplaces','Mas Vnr Area','Exter Qual','SalePrice']

df = df[variables_finales]
df
```

```python
df.info()
```

Neighborhood --> dummies
Kitchen Qual --> one-hot
Foundation --> one-hot
Exter Qual --> one-hot

```python
	#foundation
valores_foundation = {'PConc': 1, 'CBlock': 2, 'BrkTil': 3, 'Slab': 4, 'Stone': 5, 'Wood': 6}
df['Foundation'] = df['Foundation'].replace(valores_foundation)
df['Foundation']
```

```python
col_qual = ['Kitchen Qual', 'Exter Qual']

#Kitchen Qual y Exter Qual
valores_qual = {'TA': 1, 'Gd': 2, 'Ex': 3, 'Fa': 4, 'Po': 5}

df[col_qual] = df[col_qual].replace(valores_qual)


#Neighborhood
df = pd.get_dummies(df, columns=['Neighborhood'], drop_first=True)
df.info()
```

```python
# Lista de todas las columnas booleanas para convertir a INT
columnas_bool = df.select_dtypes(include='bool').columns
df[columnas_bool] = df[columnas_bool].astype(int)

df.info()
df = df.fillna(0)
```

## Entrenamiento de Modelos

#### Regresion Lineal

```python
import numpy as np

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score


#Definicion de variables
X = df.drop('SalePrice', axis=1)
y = df['SalePrice']


#Conjuntos de entrenamiento
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

print(f"Nmero de caractersticas (k): {X_train.shape[1]}")
print(f"Nmero de observaciones de prueba (n): {X_test.shape[0]}")
```

```python
#defino y entreno el modelo
model = LinearRegression()
model.fit(X_train, y_train)

#predicciones
y_pred = model.predict(X_test)
```

```python
# 3. Clculo de Mtricas de Error
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)

print("--- Mtricas de Error (Unidad: Cientos de Miles de Dlares) ---")
print(f"MAE (Error Absoluto Medio): {mae:.4f}")
print(f"MSE (Error Cuadrtico Medio): {mse:.4f}")
print(f"RMSE (Raz del Error Cuadrtico Medio): {rmse:.4f}")
```

MAE ($20,181): De media, tu modelo se equivoca por unos 20 mil dlares. Es una cifra decente para un primer modelo.

RMSE ($32,063): Fjate que el RMSE es mucho mayor que el MAE. Esto te est gritando que tienes outliers (valores atpicos). El RMSE penaliza los errores grandes de forma cuadrtica; si tienes un par de casas donde el modelo fall por $100,000, el RMSE subir muchsimo aunque el MAE sea bajo.

```python
#r2score
rscore = r2_score(y_test,y_pred)
print(f"R2 Score del Modelo Base: {rscore:.4f}")
```

```python
import matplotlib.pyplot as plt 
import seaborn as sns

plt.figure(figsize=(10,6))

sns.scatterplot(x = y_test, y=y_pred, alpha=0.5)

linea_prediccion_per = [y_test.min(), y_test.max()]
plt.plot(linea_prediccion_per, linea_prediccion_per, color='red', linestyle='--')

plt.title('Valores Reales vs. Predicciones (Ames Housing)')
plt.xlabel('Precio Real')
plt.ylabel('Precio Prediccion')
plt.show()
```
